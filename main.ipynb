```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


def plot_distribution(data, column, plot_type='hist', height=8, aspect=None):
    if plot_type == 'hist':
        sns.displot(data=data, x=column, kde=True, height=height, aspect=aspect)
    elif plot_type == 'count':
        sns.catplot(data=data, x=column, kind='count', height=height, aspect=aspect)
    plt.title(f'Distribution of {column}')
    plt.show()

def describe_column(data, column):
    unique_values = data[column].nunique()
    value_counts = data[column].value_counts()
    
    print(f"Column: {column} has {unique_values} unique values\n")
    print(value_counts)

def plot_categorical_distribution(data: pd.DataFrame = None, column: str = None, height: int = 8, aspect: int = 2):
    sns.catplot(data=data, x=column, kind='count', height=height, aspect=aspect)
    plt.title(f'Distribution of {column}')
    plt.show()

# Example usage with your specific column names:
plot_distribution(data, 'category', plot_type='hist', height=8, aspect=2)
describe_column(data, 'category')
plot_categorical_distribution(data,'category')
```

```python
plot_distribution(data, 'unit_price', plot_type='hist', height=8, aspect=2)
describe_column(data, 'unit_price')
plot_categorical_distribution(data,'unit_price')
```

```python
def convert_to_datetime(data: pd.DataFrame = None, column: str = None):
    waqt = data.copy()
    waqt[column] = pd.to_datetime(waqt[column], format='%Y-%m-%d %H:%M:%S')
    return waqt
```

```python
data = convert_to_datetime(data, 'timestamp')
```

```python
data.info()
```

```python
data['hour'] = data['timestamp'].dt.hour
```
